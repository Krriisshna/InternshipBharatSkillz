```python
#Color each dot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","deeppink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x,y, c = colors)

plt.show()
```

```python
#Colormap
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0,10,23,45,32,67,54,87,99,100,33,43,27])

plt.scatter(x,y, c = colors, cmap = 'Accent_r')

plt.show()
```

```python
#Colorbar
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y, c = colors, cmap = 'afmhot')

plt.colorbar()

plt.show()
```

```python
#Size of dot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y, s = sizes)

plt.show()
```

```python
#Transparency of dots
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y, s = sizes, alpha = 0.5)

plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)#adding Shadow
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "deeppink", "b", "#4CAF50"]

plt.pie(y,labels=mylabels,colors=mycolors)#adding color for each wedge
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y,labels=mylabels)
plt.legend()#explanation for each wedge
plt.show()
```
