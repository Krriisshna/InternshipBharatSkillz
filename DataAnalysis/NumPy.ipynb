```python
#NumPy(Numerical Pythond)
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr))
```

```python
print(np.__version__)
```

```python
arr = np.array((1,2,3,4,5))#we can pass any array like object to make ndarray like list,tuple,etc.

print(arr)
```

```python
#Dimensions in array:
arr = np.array(42) #0-D array or Scalars
print(arr)
```

```python
arr = np.array([1,2,3,4,5]) #1-D array or uni-dimensional array which have 0-D arrays as elements.
print(arr)
```

```python
arr = np.array([[1,2,3],[4,5,6]]) 
#2-D array which have 1-D arrays as elements.
print(arr)
```

```python
arr = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
#3-D array which have 2-D arrays as elements.
print(arr)
```

```python
#ndim attribute that returns no. of dimensions
a = np.array(42)
b = np.array([1,2,3,4,5])
c = np.array([[1,2,3],[4,5,6]])
d = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```
