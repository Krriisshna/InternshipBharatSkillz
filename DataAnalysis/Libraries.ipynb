```python
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Number of rows
n = 100

# Generate fake data
data = {
    'Employee_ID': np.arange(1, n+1),
    'Name': [f'Employee_{i}' for i in range(1, n+1)],
    'Age': np.random.randint(21, 60, n),
    'Department': np.random.choice(['HR', 'IT', 'Finance', 'Sales', 'Marketing'], n),
    'Experience_Years': np.random.randint(0, 20, n),
    'Salary': np.random.randint(25000, 120000, n),
    'Joining_Date': pd.to_datetime(np.random.choice(pd.date_range('2015-01-01', '2023-12-31'), n))
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV for later use
df.to_csv("employee_data.csv", index=False)

dtale.show(df.head())
```

```python
df = pd.read_csv('workout_data.csv')
print(df)
```

```python
#Removing rows containing null values
new_df = df.dropna()
dtale.show(new_df)
```

```python
#dropna() will return a new dataframe so we use argument inplace=True to change the original df
df.dropna(inplace=True)
dtale.show(df)
```

```python
df = pd.read_csv('workout_data.csv')
df.fillna(130,inplace=True)
dtale.show(df)
```

```python
df = pd.read_csv('workout_data.csv')
df.fillna({"Calories": 130}, inplace=True)
dtale.show(df)
```

```python
df = pd.read_csv('workout_data.csv')
x = df["Calories"].mean()
df.fillna({"Calories": x}, inplace=True)
dtale.show(df)
```

```python
df = pd.read_csv('workout_data.csv')
x = df["Calories"].median()
df.fillna({"Calories": x}, inplace=True)
print(x)
dtale.show(df)
```

```python
df = pd.read_csv("workout_data.csv")
x = df["Calories"].mode()[0]
df.fillna({"Calories": x}, inplace=True)
print(x)
dtale.show(df)
```

```python
#Handling wrong data format
df = pd.read_csv('workout_data.csv')
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
dtale.show(df)
```

```python
df.dropna(subset=["Date"],inplace=True)
dtale.show(df)
```

```python
#Fixing wrong data
df.loc[7, 'Duration'] = 45
dtale.show(df)
```

```python
#Setting conditions to replace values on wrong data cells
df.loc[7, 'Duration'] = 450
df.dropna(inplace=True)
for x in df.index:
    if df.loc[x,"Duration"] > 120:
        df.loc[x, "Duration"] = 120
dtale.show(df)
```

```python
#Setting conditions to drop rows on wrong data cells
df.loc[7, "Duration"] = 450
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace=True)
dtale.show(df)
```

```python
#Handling duplicate values
print(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
dtale.show(df)
```
