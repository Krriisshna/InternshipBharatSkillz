```python
#pyplot submodule
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xpoints = np.array([0,6])
ypoints = np.array([0,250])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,8])
ypoints = np.array([3,10])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
ypoints = np.array([3,8,1,10,5,7])
plt.plot(ypoints)
plt.show()
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(ypoints, marker = 'D')
plt.show()
```

```python
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
plt.plot(ypoints, '*--c')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', markersize = 20) #ms(markersize)
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r') #mec(markeredgecolor)
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r') #mfc(markerfacecolor)
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'deeppink', mfc = 'deeppink')
plt.show()
```
